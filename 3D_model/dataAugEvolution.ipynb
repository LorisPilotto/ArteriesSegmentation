```python
import matplotlib.pyplot as plt
import numpy as np
```

# 3D simple UNet

```python
t1 = np.load('IoU/3D_2_1_50EPOCH_simple_model_Train.npy')
v1 = np.load('IoU/3D_2_1_50EPOCH_simple_model_Val.npy')
```

```python
epochs = range(50)

fig, ax = plt.subplots(1, 1, figsize=(15,7))
#metric: IoU
ax.plot(epochs, t1, 'r', label='2x512x512 Training IoU', color='red')
ax.plot(epochs, v1, 'bo', color='red')
ax.plot(epochs, v1, '--', label='2x512x512 Validation IoU', color='red')

"""ax.plot(epochs, t2, 'r', label='Training IoU with drop layer', color='blue')
ax.plot(epochs, v2, 'bo', color='blue')
ax.plot(epochs, v2, '--', label='Validation IoU with drop layer', color='blue')

ax.plot(epochs, t3, 'r', label='Training IoU with drop layer & data augmentation', color='green')
ax.plot(epochs, v3, 'bo', color='green')
ax.plot(epochs, v3, '--', label='Validation IoU with drop layer & data augmentation', color='green')"""

ax.set_title('3D model with 2x512x512 blocks:')
ax.set(xlabel='Epoch', ylabel='IoU Value', ylim=[0,1])
ax.legend()
plt.show()
```

# Models combined & data augmentation:

```python
combined_t1 = np.load('IoU/models_combined_drop_aug_3D_2_1_50EPOCH_Train.npy')
combined_v1 = np.load('IoU/models_combined_drop_aug_3D_2_1_50EPOCH_Val.npy')
```

# 4x512x512 blocks:

```python
combined_t2 = np.load('IoU/models_combined_drop_aug_3D_4_1_50EPOCH_Train.npy')
combined_v2 = np.load('IoU/models_combined_drop_aug_3D_4_1_50EPOCH_Val.npy')
```

```python
epochs = range(50)

fig, ax = plt.subplots(1, 1, figsize=(15,7))
#metric: IoU
ax.plot(epochs, t1, 'r', label='2x512x512 Training IoU', color='red')
ax.plot(epochs, v1, 'bo', color='red')
ax.plot(epochs, v1, '--', label='2x512x512 Validation IoU', color='red')

ax.plot(epochs, combined_t1, 'r', label='2x512x512 Training IoU with data augmentation', color='blue')
ax.plot(epochs, combined_v1, 'bo', color='blue')
ax.plot(epochs, combined_v1, '--', label='2x512x512 Validation IoU with data augmentation', color='blue')

ax.plot(epochs, combined_t2, 'r', label='4x512x512 Training IoU with data augmentation', color='green')
ax.plot(epochs, combined_v2, 'bo', color='green')
ax.plot(epochs, combined_v2, '--', label='4x512x512 Validation IoU with data augmentation', color='green')

ax.set_title('3D models:')
ax.set(xlabel='Epoch', ylabel='IoU Value', ylim=[0,1])
ax.legend()
plt.show()
```
