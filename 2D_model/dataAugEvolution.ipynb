```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
print(v1[25:40].mean())
print(v2[25:40].mean())
print(v3[25:40].mean())
```

```python
epochs = range(40)

fig, ax = plt.subplots(1, 1, figsize=(15,7))
#metric: IoU
ax.plot(epochs, t1, 'r', label='Training IoU no aug', color='red')
ax.plot(epochs, v1, 'bo', color='red')
ax.plot(epochs, v1, '--', label='Validation IoU no aug', color='red')

ax.plot(epochs, t2, 'r', label='Training IoU with drop layer', color='blue')
ax.plot(epochs, v2, 'bo', color='blue')
ax.plot(epochs, v2, '--', label='Validation IoU with drop layer', color='blue')

ax.plot(epochs, t3, 'r', label='Training IoU with drop layer & data augmentation', color='green')
ax.plot(epochs, v3, 'bo', color='green')
ax.plot(epochs, v3, '--', label='Validation IoU with drop layer & data augmentation', color='green')

ax.plot(epochs, combined_t3, 'r', label='Training IoU with drop layer & data augmentation', color='purple')
ax.plot(epochs, combined_v3, 'bo', color='purple')
ax.plot(epochs, combined_v3, '--', label='Validation IoU with drop layer & data augmentation', color='purple')

ax.set_title('Training and Validation IoU (the metric):')
ax.set(xlabel='Epoch', ylabel='IoU Value', ylim=[0,1])
ax.legend()
plt.show()
```
